```python
import pandas as pd
import numpy as np
import re
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from google.colab import files


uploaded = files.upload()
```

```python
# Load the dataset from the CSV file
data = pd.read_csv('ph.csv')
```

```python
# Extract the URLs from the dataset
urls = data['url'].values
print(urls)
```

```python
# Preprocess the URLs by tokenizing and padding them to a fixed length
tokenizer = Tokenizer()
tokenizer.fit_on_texts(urls)
sequences = tokenizer.texts_to_sequences(urls)
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')
print(padded_sequences)
```

```python
# Split the dataset into training and testing sets
labels = data['label'].values
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

```

```python
 
model = Sequential()
model.add(Embedding(len(tokenizer.word_index) + 1, 64, input_length=max_length))
model.add(LSTM(64, dropout=0.2, recurrent_dropout=0.2))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the LSTM model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10, batch_size=32)
```

```python
# Use the trained LSTM model to predict the labels of the URLs in the dataset
y_pred = model.predict(padded_sequences)
```

```python
# Extract the hyperlinks from the phishing URLs and print them
for i in range(len(urls)):
    if y_pred[i] > 0.5:
        match = re.findall(r'(https?://\S+)', urls[i])
        if match:
            print(f"Phishing URL: {urls[i]}")
            print(f"Hyperlink: {match[0]}\n")
```
